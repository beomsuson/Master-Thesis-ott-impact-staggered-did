```python
import pandas as pd

# 1. 데이터 불러오기
df = pd.read_csv("movie_long_with_emotions_full.csv")

# 2. 날짜 변환
df["개봉일"] = pd.to_datetime(df["개봉일"], errors="coerce")
df["일자"] = pd.to_datetime(df["일자"], errors="coerce")

# 3. 분석 기간 설정
start_date = "2023-01-01"
end_date = "2024-12-31"

# 4. 대상 영화: 개봉일이 2023~2024년인 영화
df = df[(df["개봉일"] >= start_date) & (df["개봉일"] <= end_date)]

# 5. DATE 패널 생성 (2023-01-01 ~ 2024-12-31)
date_range = pd.date_range(start=start_date, end=end_date, freq="D")
date_df = pd.DataFrame({"DATE": date_range})

# 6. 영화명 × DATE 패널 확장
movies = df["영화명"].unique()
panel = pd.MultiIndex.from_product([movies, date_range], names=["영화명", "DATE"]).to_frame(index=False)

# 7. 원본 데이터와 merge (DATE = 일자)
df_merged = panel.merge(df, left_on=["영화명", "DATE"], right_on=["영화명", "일자"], how="left")

# 8. 결과 저장
df_merged.to_csv("movie_long_panel_2023_2024.csv", index=False, encoding="utf-8-sig")
print("✅ 최종 패널 저장 완료 → movie_long_panel_2023_2024.csv")
print(df_merged.head(10))
```
